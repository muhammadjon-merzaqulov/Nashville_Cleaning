```python
import pandas as pd
```

```python
df = pd.read_excel('/content/Nashville.xlsx')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# PropertyAddress dagi null qiymatlarni tekshirish
df[df['PropertyAddress'].isnull()].head()
```

```python
# ParcelID bo'yicha birinchi mavjud PropertyAddress qiymatini olish va null qiymatlarni olib tashlash
property_map = df.groupby('ParcelID')['PropertyAddress'].first().dropna()

# Null qiymatlarni to'ldirish uchun .map() dan foydalaning
df['PropertyAddress'] = df['PropertyAddress'].fillna(df['ParcelID'].map(property_map))
df.head()
```

```python
# null qiymat bor yo'qligini qayta tekshirish
df[df['PropertyAddress'].isnull()].head()
```

```python
# 1. PropertyAdressdan uy nomerini ajratib olish
df['P_A_num'] = df['PropertyAddress'].str.extract(r'^(\d+)')

# 2. PropertyAdressdan ko'chani ajratib olish
df['P_A_street'] = df['PropertyAddress'].str.replace(",", "").str.split(" ", n=2).str[1:].str.join(" ")

# 3. PropertyAdressdan cityni ajratib olish
df['P_A_city'] = df['PropertyAddress'].str.split(",").str[-1].str.strip()
df.sample(10)
```

```python
# PropertyAddress tashlab yuborish va df chiqarish
df.drop('PropertyAddress', axis=1, inplace=True)
df.head()
```

```python
# 1. OwnerAddress dan uy nomerini ajratib olish
df['Owner_A_num'] = df['OwnerAddress'].str.extract(r'^(\d+)')

# 2. OwnerAddress dan ko'chani ajratib olish
df['Owner_A_street'] = df['OwnerAddress'].str.extract(r'^\d+\s+([A-Za-z\s]+)')

# 3. OwnerAddress dan cityni ajratib olish
df['Owner_A_city'] = df['OwnerAddress'].str.split(",").str[1].str.strip()
df.sample(10)

# 4. OwnerAddress dan state ni ajratib olish
df['Owner_A_state'] = df['OwnerAddress'].str.split(",").str[-1].str.strip()

# Tasodifiy 10 qatorni chiqarish
df.sample(10)
```

```python
# OwnerAddress tashlab yuborish va df chiqarish
df.drop('OwnerAddress', axis=1, inplace=True)
df.head()
```

```python
# SoldAsVacant Y | N qiymatlar borligini tekshirish
df['SoldAsVacant'].value_counts()
```

```python
# SoldAsVacant Y | N ni --> Yes | No o'zgartirish
df['SoldAsVacant'] = df['SoldAsVacant'].replace({'Y': 'Yes', 'N':'No'})
```

```python
# SoldAsVacant Y | N qiymatlar borligini tekshirish
df['SoldAsVacant'].value_counts()
```

```python
df.shape
```

```python
df.drop_duplicates(subset=['ParcelID', 'LandUse', 'SaleDate', 'SalePrice', 'LegalReference', 'SoldAsVacant', 'OwnerName',
                           'Acreage', 'TaxDistrict', 'LandValue', 'BuildingValue', 'TotalValue', 'YearBuilt', 'Bedrooms',
                           'FullBath', 'HalfBath', 'P_A_num', 'P_A_street', 'P_A_city', 'Owner_A_num', 'Owner_A_street',
                           'Owner_A_city', 'Owner_A_state'], inplace=True)
df.shape
```

```python
# OwnerName da null qiymatlarni soni
df['OwnerName'].isnull().sum()
```

```python
# OwnerName null qiymatlarni Unknown o'zgartirish
df['OwnerName'] = df['OwnerName'].fillna('Unknown')

# OwnerName null qiymatlarni sonini qayta tekshirish
df['OwnerName'].isnull().sum()
```

```python
df.to_excel('Nashville_new.xlsx', index=False)
```
